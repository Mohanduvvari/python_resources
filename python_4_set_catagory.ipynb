```python
#add(Value)
#add an item to to the set
s=set(["mohan",123,89,9090])
print(s,len(s),type(s),id(s))
print('-'*50)
s.add("rammohan reddy")
print(s,len(s),type(s),id(s))
print('-'*50)
```

```python
#remove(value)
#remove an value form set
s=set(["mohan",123,89,9090,"rammohan"])
print(s,len(s),type(s),id(s))
print('-'*50)
s.remove(89)
print(s,len(s),type(s),id(s))
print('-'*50)
s.remove(10000) #gives key error as 10000 not in set
print(s,len(s),type(s),id(s))
print('-'*50)

```

```python
#discard(value)
#if the value not in list does not gives any error like remove()
s=set(["mohan",123,89,9090,"rammohan"])
print(s,len(s),type(s),id(s))
print('-'*50)
s.discard(89)
print(s,len(s),type(s),id(s))
print('-'*50)
s.discard(10000) #does not gives error insted gives none or just space as result in python ides not in realtime ides
print(s,len(s),type(s),id(s))
print('-'*50)

```

```python
#pop()
s={9090, 'rammohan', 89, 123, 'mohan'}
print(s,len(s),type(s),id(s))
print('-'*50)
s.pop() #pop a item randomly
print(s,len(s),type(s),id(s))
```

```python
#clear() #clears entire set gives empty set
s={9090, 'rammohan', 89, 123, 'mohan'}
print(s,len(s),type(s),id(s))
print('-'*50)
s.clear() 
print(s,len(s),type(s),id(s))
del s
print(s,len(s),type(s),id(s))
```

```python
#isdisjoin() #returns true if there is atleast one common item,else false
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set(["9090",99,123])
s3=set((1231,453))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s2.isdisjoint(s1))
print(s3.isdisjoint(s1))
print(set().isdisjoint(s1))
```

```python
#issubset()
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set(["9090",99,123])
s3=set((9090,123))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s2.issubset(s1))
print(s3.issubset(s1))
print(set().issubset(s1))
```

```python
#issuperset()
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set(["9090",99,123])
s3=set((9090,123))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s1.issuperset(s2))
print(s1.issuperset(s3))
```

```python
#union() 
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set([9090,99,123])
s3=set((9090,123))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s2.union(s3))
print('-'*50)
print(s1|s2)  #Bit Wise OR (|) 
print('-'*50)
print(s1.union(s2)) 
```

```python
#intersection
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set([9090,99,123])
# s3=set((9090,123))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s2.intersection(s1))
print('-'*50)
print(s2&s1) #bitwise AND(&)
```

```python
#difference (minu - )
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set([9090,99,123])
# s3=set((9090,123))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s2.difference(s1))
print('-'*50)
print(s2-s1) # - 
print('-'*50)
print(s1-s2)
```

```python
#symmetric_difference() 
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set([9090,99,123])
# s3=set((9090,123))
print(s1,len(s1),type(s1),id(s1))
print(s2,len(s2),type(s2),id(s2))
print('-'*50)
print(s2.symmetric_difference(s1))
print('-'*50)
print(s2^s1) # bitwise XOR (^)
print('-'*50)
print(s1^s2)
```

```python
#update() 
s1={9090, 'rammohan', 89, 123, 'mohan'}
s2=set([9090,99,123])
s3=set((9090,123,'raju'))
print(s1,len(s1),type(s1),id(s1))
print(s3,len(s3),type(s3),id(s3))
print('-'*50)
s3.update(s1)
print(s3,len(s3),id(s3))
print('-'*50)
#from now onwards s3 becomes  {9090, 'mohan', 'rammohan', 'raju', 89, 123}
```

```python
#similar for intersection_update
#similar for difference_update
#similar for symmetric_difference_update
```

```python
#intersection_update()
a=set((12,34,54,90900,782))
b=set((11,33,44,55,66,22,12))
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
print("-"*50)
a.intersection_update(b)
print(a,type(a),len(a),id(a))
#from now onwards a becomes 12
```

```python
#difference_update
a=set((12,34,54,90900,782))
b=set((11,33,44,55,66,22,12))
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
print("-"*50)
a.difference_update(b)
print(a,type(a),len(a),id(a))
```

```python
#symmetric_difference_update
a=set((12,34,54,90900,782))
b=set((11,33,44,55,66,22,12))
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
print("-"*50)
a.symmetric_difference_update(b)
print(a,type(a),len(a),id(a))
```

```python
#difference between difference and symmetric difference
a={10,20,30,40}
b={20,30,50,60}
print('difference=',a-b)
print('symmetric_difference=',a^b)
```

```python
#frozen set()
#frozen set is purely immutable
#allows only unique values
#elements of frozen set must be obtained from objects like set,tuple,list..
#frozen set possess all characters of set but not support set predefine functions
#frozensetobj=frozenset(set,tuple,list,str,bytes,bytearray,range)
```

```python
#empty frozenset
a=frozenset()
print(a,type(a),len(a))
```

```python
b=frozenset({12,34,56,45,22}) #set
c=frozenset([11,22,33,44,11,45,67,98,64]) #list 
d=frozenset(('m','k','j','d'))  #tuple
e=frozenset("mohankrishna") #str
f=frozenset(range(1,10)) #range
#also we can give bytes, bytearray
```

```python
print(b,type(b),len(b))
print(c,type(c),len(c))
print(d,type(d),len(d))
print(e,type(e),len(e))
print(f,type(f),len(f))
```
