```python
# bytes
#immutable datatype (data cannot be manipulated at same memory location),does not allow item assignment
#follows insertion order
# we can perform indexing and slicing on the bytes data type
#bytes can be implemented on iterable objects
#bytes can be implemented on the range of values (0-256)
#used tp perform end-to-end encryption over remote devices
```

```python
# a=['a','b','c','d'] #iterable but cannot convert into bytes
# b=['a',12,236,True,12.90,3+4j] #iterable but cannot convert into bytes
# c=[125,65,96,74,85,100,256,-34] #iterable but cant convert into bytes (0-255)giving 256,-34
# # print(bytes(a))
# # print(bytes(b))
# # print(bytes(c))
```

```python
a=[False,True,1,56]#here bool values can be allows coz internally they can be represented as 0,1
bytes(a)
```

```python
import sys
a=[0,25,63,45,123,255]
print(a,",",type(a),",",id(a),",",sys.getsizeof(a),"bytes",",",'length:',len(a))
print("-"*50)
b=bytes(a)
print(b,",",type(b),",",id(b),",",sys.getsizeof(b),"bytes",",",'length:',len(a))
```

```python
#bytes allow indexing
print(b[0],b[1],b[2]) #accessing bytes object data using indexing
print(b[0:4]) # calling first 4 values using slicing from bytes object
```

```python
#bytes does not allow item assignment
print(b[1])

b[1]=56 #item assignment not allowed
print(b[1])
```

```python
#accessing items from bytes object (use looping)

print('accessing all items from object')
for i in b:
       print(i,end=" ")
print('\n')
print('accessing 5 items from object')
for i in b[:5]:
        print(i,end=" ")
print('\n')
print("WE can also perform memory managment techniques on the bytes object")
for i in b[::-1]: #b[0::2]....etc
        print(i,end=" ")

```

```python
#int()
#Used to convert any possible value to int type

#varname=int(float/bool/complex/str)

#float to int possible
x=236.52
try:
    fi=int(x) #here the decimal was places were removed
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

#bool to int possible
x=True
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

#complex to int impossible
x=3+4j
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

#str to int 
#str has str_int,str_float,str_bool,str_complex,pure_str
x='567' #str_int to int possible
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

x='695.354' #str_float to int not possible
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

x='False' #str_bool to int not possible
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

x='5+7j' #str_complex to int not possible
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)

x='mohan' #purestr to int not possible
try:
    fi=int(x) 
    print(f"{type(x)} to {type(fi)} possible ")
    print(f"{x} is converted to {fi} using int() ")
    print(f"type of '{x}' is {type(x)}\ntype of '{fi}' is {type(fi)}")
except Exception:
    print(f"{type(x)} to {type(fi)} not possible ")
print('-'*70)
```
