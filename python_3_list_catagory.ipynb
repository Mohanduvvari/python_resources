```python
#copy -copy items from one lst obj to another list obj
#1.1 shallow copy  -content is same but memory address is diff modification is      independent - which changes we done on one obj will not reflect to other obj  
#1.2 deep copy  -content is same and memory is also same, modification is dependent
```

```python
#shallow copy
#copy()
a=[12,34,56,78,90,12,222]
b=a.copy()
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
print('-'*50)
a.pop()
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
```

```python
#deep copy
a=[12,34,56,78,90,12,222,'mahesh']
b=a #deep copy
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
print('-'*50)
a.insert(2,'mallesh')
print(a,type(a),len(a),id(a))
print(b,type(b),len(b),id(b))
```
