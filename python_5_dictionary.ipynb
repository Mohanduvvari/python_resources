```python
#Dictionary data type (MAP)
#purpose is to store data in key,value pair
#keys must be unique, but values can be duplicate 
#dictobj maintain insertion order- ordered
#keys of dictobj are immutable
#values for dict obj is mutable 
#dictobj never have indexes so wen can't access keys with indexs
#we can access values by using keys of dictobj
#dict does not support indexng and slicing
```

```python
#we can create two types of dict objes 
#1.empty dictobj--> {},dict()
#2non_empty dicobj
```

```python
#empty dictionary
d1={}
print(d1,type(d1),len(d1),id(d1))
print('-'*50)
```

```python
d1[1]=234 #here 1 is not index it is key and 234 is value for key
d1
```

```python
d1[3]=456
d1[4]=9090
d1[6]='mk'
print(d1,type(d1),len(d1),id(d1)) 
#here 1,3,4,6 are keys
```

```python
#non-empty dict
a={'apple':20,'banana':90,'cherry':50,'dryfruits':20}
print(a)
print('-'*40)

#accessing using keys
print(a['apple'])
print(a['dryfruits'])
# print(a['zebra']) #gives key error 
```

```python
#replace the value
a['apple']=100 #previously apple 20 now it becomes 100
print(a)
```

```python
#key duplicates
b={10:'ten',20:'twenty',30:'thirty',10:'bten'} #here key 10 was duplicated so its value becomes last key value -->ten becoms bten
print(b)
```

```python
#value duplication
b={10:'ten',20:'twenty',30:'thirty',40:'ten',60:'ten'} #here value was duplicated 
print(b)
```

```python
#creating a new dictionary with keys for a itrobj
a="hello"
nd=dict.fromkeys(a,0) #here a is itrobj, 0 is default
nd
```

```python
 #predefined funtions on dictionary
# 1.clear()
# 2.pop()
# 3.popitem()
# 4.copy()
# 5.get()
# 6.keys()
# 7.values()
# 8.items()
```

```python
#clear()
b={10:'ten',20:'twenty',30:'thirty',40:'fourty',60:'sixty'} 
print(b,type(b),len(b),id(b))
print('-'*50)
b.clear()
print(b,type(b),len(b),id(b))
print('-'*50)
print({}.clear()) #coz it is already emtpy
```

```python
#pop(key)
b={10:'ten',20:'twenty',30:'thirty',40:'fourty',60:'sixty'} 
print(b,type(b),len(b),id(b))
print('-'*50)
b.pop(40)
print(b,type(b),len(b),id(b))
```

```python
#popitem() remove last item
b={10:'ten',20:'twenty',30:'thirty',40:'fourty',60:'sixty'} 
print(b,type(b),len(b),id(b))
print('-'*50)
b.popitem()
print(b,type(b),len(b),id(b))
```

```python
#copy() shallow copy
b={10:'ten',20:'twenty',30:'thirty',40:'fourty',60:'sixty'} 
print(b,type(b),len(b),id(b))
print('-'*50)
d=b.copy()
print(d,type(d),len(d),id(d))
```

```python
c=d
print(id(c),id(d))
c[34]='thirty four'
print(c)
print(d)
```

```python
#get(key)
# c[343] gives key error
print(c.get(343)) #gives none mean nothing is there with 343
print(c.get(34))
```

```python
#keys() gives all keys from dict as dict_keys class
print(c.keys())
print('-'*50)
for i in c.keys():
    print(i)
```

```python
#values() gives all values from dict as ordered with dict_values class
print(c.values())
print('-'*50)
for i in c.values():
    print(i)
```

```python
for k in {10: 'ten', 20: 'twenty', 30: 'thirty', 40: 'fourty', 60: 'sixty', 34: 'thirty four'}.keys():
    print(k,end=" ")
#we can also use enumerate funtion her
```

```python
for k in {10: 'ten', 20: 'twenty', 30: 'thirty', 40: 'fourty', 60: 'sixty', 34: 'thirty four'}:
    print(k,end=" ")
```

```python
for k in enumerate({10: 'ten', 20: 'twenty', 30: 'thirty', 40: 'fourty', 60: 'sixty', 34: 'thirty four'}):
    print(k)
```

```python
b={10:'ten',20:'twenty',30:'thirty',40:'fourty',60:'sixty'}
for k in enumerate(b):
    print(k,'-->',b[k[1]]) #here we access k's 1st value
```

```python
#items()
b={10:'ten',20:'twenty',30:'thirty',40:'fourty',60:'sixty'}
print(b.items())
for i,j in b.items():
    print(i,'-->',j)
```

```python
for v in enumerate(b):
    print(v,'-->',b[v[1]])
```
