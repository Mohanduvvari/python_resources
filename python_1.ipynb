```python
# int to decimal number system
#bin,oct,hex
a=23
b=bin(a)
c=oct(a)
d=hex(a)
print(f"a:{a}->{type(a)}, binary form is {b}-->{type(b)}")
print(f"\t\t\toctal form is {c}-->{type(c)}")
print(f"\t\t\thexa decimal form is {d}-->{type(d)}")

#the conversion give the form is 'str' type
```

```python
# Decimal number system to int
#base of decimal number system is - 10
#base of binary number system is - 2
#base of octal number system is - 8
#base of hexa decimal number system is - 16
a=23 #decimal form
b=bin(a) #0b10111
c=oct(a) #0o27
d=hex(a) #0o27

#to convert from bin/octal/hexa to decimal we use fomat as:

e=int(b,2) #binary to decimal we use use binary base (2)
f=int(c,8) #octal to decimal we use its octal base  (8)
g=int(d,16) #hexa to decimal we use its hexa base (16)


print(f"{b}-->{e},{type(e)}")
print(f"{c}-->{f},{type(f)}")
print(f"{d}-->{g},{type(g)}")

```

```python
#Decimal numbersystem advance
a=1025 #decimal form
b=bin(a) #0b10111
c=oct(a) #0o27
d=hex(a) #0o27
print(f"binary of {a} is {a:b}") #without prefix 0b
print(f"binary of {a} is {a:#b}") #with prefix 0b
print(f"octal of {a} is {a:o}")
print(f"octal of {a} is {a:#o}") #with prefix 0o
print(f"hexa of {a} is {a:x}")
print(f"hexa of {a} is {a:#X}")
```

```python
aa=77

print(f"{aa}-->{aa:x}") # small x is for lower character
print(f"{aa}-->{aa:#x}")

print(f"{aa}-->{aa:X}") # upper X is for upper characters
print(f"{aa}-->{aa:#X}")
```
